# Geemap LiDAR Viewer

This is a demonstration of how to use Geemap to view LiDAR point clouds. For starters, Geemap is a Python package that allows for interactive mapping with the Google Earth Engine. In other words, it enables users to interactively visualize and analyze GEE datasets.

For this demonstration, I will use a LAS file from an area near Waupaca, Wisconsin in close proximity to both Mirror and Shadow Lake. You may use any Earth Engine data file format of your choice assuming it reflects the area you desire to analyze for your project.

# Loading the LAS file

```python
file = r'C:\projects\waupaca.las'
```

# Reading the LAS file

```python
lasfile = geemap.read_lidar(file)
```

```python
lasfile.header
```

```python
lasfile.header.point_count
```

```python
list(lasfile.point_format.dimension_names)
```

```python
lasfile.X
```

```python
lasfile.intensity
```

# View the Elevations Using Geemap
Here we can see the full elevations for the area represented by our LAS file.

```python
geemap.view_lidar(file, cmap='terrain', backend='pyvista')
```

# Changing the Colormaps
Additionally, you may also change the colormaps of your visuals to suit your own individual preferences. The above visual used the 'terrain' colormap, while the following two use the 'viridis' and 'magma' colormaps, respectively.

```python
geemap.view_lidar(file, cmap='viridis', backend='pyvista')
```

```python
geemap.view_lidar(file, cmap='magma', backend='pyvista')
```
